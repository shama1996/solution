# The task assigned to you is to scrape archives from The Times of India website for years 2015 to 2021. You will be allotted tasks to scrape data based on different timeframes. The details to be scraped are given below: 

1. Date: This column will contain the date on which news was reported. 
2. Author: This column will contain name of the author of news report. 
3. Vertical: This column will contain the news verticals (e.g. Entertainment, Sports, Lifestyle, Economy, etc.). 
4. Headline: This column will contain the news headline. 
5. Description: This column will contain the main content of the news report. 
You have to write a code to scrape the above data from The Times of India website

```python
#importing all the required libraries
import selenium
import pandas as pd
from selenium import webdriver # Importing selenium webdriver 

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
driver=webdriver.Chrome(executable_path="C:\\Users\\HP\\chromedriver.exe")
```

```python
# Opening the url
driver.get('https://timesofindia.indiatimes.com/archive/year-2017,month-12.cms')
```

```python
#creating empty lists
Date=[]
Author=[]
Vertical=[]
Headline=[]
Description=[]
```

```python
#creating a list url so that all the urls of date from 1st december to 31st december can be appended
url=[]
for i in driver.find_elements_by_xpath("//div[@id='calenderdiv']/table/tbody/tr[2]/td[6]/a"):
    url.append(i.get_attribute("href"))
for i in driver.find_elements_by_xpath("//div[@id='calenderdiv']/table/tbody/tr[2]/td[7]/a"):
    url.append(i.get_attribute("href"))
for i in driver.find_elements_by_xpath("//div[@id='calenderdiv']/table/tbody/tr[3]/td/a"):
    url.append(i.get_attribute("href"))
for i in driver.find_elements_by_xpath("//div[@id='calenderdiv']/table/tbody/tr[4]/td/a"):
    url.append(i.get_attribute("href"))
for i in driver.find_elements_by_xpath("//div[@id='calenderdiv']/table/tbody/tr[5]/td/a"):
    url.append(i.get_attribute("href"))
for i in driver.find_elements_by_xpath("//div[@id='calenderdiv']/table/tbody/tr[6]/td/a"):
    url.append(i.get_attribute("href"))
for i in driver.find_elements_by_xpath("//div[@id='calenderdiv']/table/tbody/tr[7]/td[1]/a"):
    url.append(i.get_attribute("href"))    
```

```python
url
```

```python
#opening the first url i.e., 1st december
driver.get(url[0])
URLS1=[] #appending all the urls of the news from 1st december
for i in driver.find_elements_by_xpath("//div[@style='font-family:arial ;font-size:12;font-weight:bold; color: #006699']/table/tbody/tr[2]/td/span/a"):
    URLS1.append(i.get_attribute("href"))
```

```python
URLS1
```

```python
#News from 1st december will be appended here with their dates,headline,vertical,description and Author
for i in URLS1[1:]:
    driver.get(i)
    try:
        date=driver.find_element_by_xpath("//div[@class='_3lvqr clearfix']/div[2]/div/div").text
        Date.append(date)
    except NoSuchElementException:
        Date.append('-')
            
    try:
        des=driver.find_element_by_xpath("//div[@class='_1_Akb clearfix  ']/div").text
        Description.append(des)
    except NoSuchElementException:
        Description.append('-')
    
    try:
        head=driver.find_element_by_xpath("//div[@class='_3lvqr clearfix']/div[2]/h1").text
        Headline.append(head)
    except NoSuchElementException:
        Headline.append('-')
    
    try:
        ver=driver.find_element_by_xpath("//li[@class='_8aEAx   K-f4t']/a").text
        Vertical.append(ver)
    except NoSuchElementException:
        Vertical.append('-')
    
    try:
        aut=driver.find_element_by_xpath("//div[@class='_3lvqr clearfix']/div[2]/div/div/span/a").text
        Author.append(aut)
    except NoSuchElementException:
        Author.append('-')
```
