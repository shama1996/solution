```python
#importing all the required libraries
import selenium
import pandas as pd
from selenium import webdriver # Importing selenium webdriver 

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
```

```python
#lets first connect to web driver
driver= webdriver.Chrome('chromedriver.exe')
#first get the webpage
driver.get('https://economictimes.indiatimes.com/archive.cms')
```

```python
dec_19=driver.find_element_by_xpath("//a[246][@class='normtxt']")
driver.get(dec_19.get_attribute("href"))
```

```python
#scraping date url
urls=driver.find_elements_by_xpath("//*[@id='calender']/tbody/tr/td/a")
URL=[]
for i in urls:
    URL.append(i.get_attribute('href'))      
```

```python
URL
```

```python
#scraping each news url 
news_urls=[]
cnt=1
for j in URL:
    driver.get(j)
    print("**************Collecting all news URL's on page ",cnt,"***********************")
    tble=driver.find_element_by_id("pageContent")
    temp=tble.find_elements_by_xpath(".//table/tbody/tr")
    for i in temp[0].find_elements_by_xpath("//ul[1][@class='content']/li/a"):
        news_urls.append(i.get_attribute('href'))
    for i in temp[2].find_elements_by_xpath("//ul[2][@class='content']/li/a"):
        news_urls.append(i.get_attribute('href'))
    print("************* All news URL's on page ",cnt, " have been stored successfully**********")
    cnt=cnt+1
```

```python
Headlines=[]
for j in URL:
    driver.get(j)
    tble=driver.find_element_by_id("pageContent")
    temp=tble.find_elements_by_xpath(".//table/tbody/tr")
    for i in temp[0].find_elements_by_xpath("//ul[1][@class='content']/li/a"):
        Headlines.append(i.text)
    for i in temp[2].find_elements_by_xpath("//ul[2][@class='content']/li/a"):
        Headlines.append(i.text)
    
```

```python
Headlines
```

```python
Author=[]
Date=[]
vertical=[]
Description=[]
```

```python


for i in news_urls:
    driver.get(i)
    WebDriverWait(driver,20)
# ********AUTHOR NAME**************    
    try:
        s=driver.find_element_by_xpath("//div[@class='auth eventDone']/a")
        Author.append(s.text)
    except NoSuchElementException as e:
        Author.append('')
#****************DATE & TIME OF UPLOAD*************
    try:
        t=driver.find_element_by_xpath("//div[@class='dtc vam artByline']/time")
        Date.append(t.text.replace('Last Updated: ',''))
    except NoSuchElementException as e:
        Date.append('')
        
#********news vertical************
    try:
        s=driver.find_element_by_xpath("//div[@class='clr breadCrumb contentwrapper']/span[3]")
        vertical.append(s.text)
    except NoSuchElementException as e:
        vertical.append('')
        
#***********Description****************************

    try:
        ds=driver.find_element_by_xpath("//div[@class='artText medium']")
        Description.append(ds.text.replace('\n\n', ''))
    
    except NoSuchElementException   as e:
        Description.append("-")
        
       
```

```python
print(len(Date),len(Author),len(vertical),len(Description))
```

```python
economictimes=pd.DataFrame({})
economictimes['Date_and_time_of_upload']=Date[:7499]
economictimes['Author_name']=Author[:7499]
economictimes['Topic_or_Vertical']=vertical[:7499]
economictimes['Headlines']=Headlines[:7499]
economictimes['Description']=Description[:7499]
economictimes
```

```python
economictimes.to_csv('economictimes_archives_december_2019.csv')
```
